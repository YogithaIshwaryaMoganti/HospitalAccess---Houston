```python
from geopy.distance import geodesic
import numpy as np
import folium
import pandas as pd
```

```python
# Define a function to calculate the distance between two points using the Haversine formula
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Radius of the earth in km
    dLat = np.deg2rad(lat2 - lat1)
    dLon = np.deg2rad(lon2 - lon1)
    a = np.sin(dLat/2) * np.sin(dLat/2) + np.cos(np.deg2rad(lat1)) * np.cos(np.deg2rad(lat2)) * np.sin(dLon/2) * np.sin(dLon/2)
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    d = R * c  # Distance in km
    return d
```

```python
centroids = pd.read_csv("centroids.csv")
hospitals = pd.read_csv("hosp.csv")
hos=pd.read_csv("hospital.csv")
```

```python
# Define the radius of the circle in miles
radius_miles = 5

# Iterate over the centroid polygons
for i, centroid_row in centroids.iterrows():
    # Get the centroid coordinates
    centroid_lat, centroid_lon = centroid_row['y'], centroid_row['x']
    # Calculate the circle center and radius in km
    circle_center = (centroid_lat, centroid_lon)
    circle_radius = radius_miles * 1.60934  # Convert miles to km
    # Initialize a count of the hospitals within the circle
    hospitals_within_circle = 0
    # Iterate over the hospitals
    for j, hosp_row in hospitals.iterrows():
        # Get the hospital coordinates
        hosp_lat, hosp_lon = hosp_row['n'], hosp_row['m']
        # Calculate the distance between the hospital and the circle center in km
        distance = haversine(centroid_lat, centroid_lon, hosp_lat, hosp_lon)
        # Check if the hospital lies within the circle
        if distance <= circle_radius:
            hospitals_within_circle += 1
    # Store the count in a new column or data structure
    centroids.loc[i, 'hospitals_within_circle'] = hospitals_within_circle
    print(f"Block {i} has {hospitals_within_circle} hospitals within {radius_miles} miles")
```

```python

# Define the radius of the circle in miles
radius_miles = 5

# Create a map centered at the first centroid point
m = folium.Map(location=[centroids.iloc[0]['y'], centroids.iloc[0]['x']], zoom_start=12)

# Iterate over the centroid polygons
for i, centroid_row in centroids.iterrows():
    # Get the centroid coordinates
    centroid_lat, centroid_lon = centroid_row['y'], centroid_row['x']
    # Calculate the circle center and radius in km
    circle_center = (centroid_lat, centroid_lon)
    circle_radius = radius_miles * 1.60934  # Convert miles to km
    # Initialize a count of the hospitals within the circle
    hospitals_within_circle = 0
    # Iterate over the hospitals
    for j, hosp_row in hospitals.iterrows():
        # Get the hospital coordinates
        hosp_lat, hosp_lon = hosp_row['n'], hosp_row['m']
        # Calculate the distance between the hospital and the circle center in km
        distance = haversine(centroid_lat, centroid_lon, hosp_lat, hosp_lon)
        # Check if the hospital lies within the circle
        if distance <= circle_radius:
            hospitals_within_circle += 1
    # Store the count in a new column or data structure
    centroids.loc[i, 'hospitals_within_circle'] = hospitals_within_circle

    # Create a circle marker with the centroid coordinates and hospitals within the circle
    folium.CircleMarker(
        location=[centroid_lat, centroid_lon],
        radius=5,
        popup=f"Block {i} has {hospitals_within_circle} hospitals within {radius_miles} miles",
        color='red',
        fill=True,
        fill_color='red'
    ).add_to(m)

# Display the map
m
```
